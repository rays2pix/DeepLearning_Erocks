### Neuron - the Building block:

Just like our neural system, neuron is the building block of an artificial neural network. A single neuron can have any number of inputs and produces one output. For a binary neuron, one could say at any instant either the neuron is activated or not activated just like a logic gate. We could extend this and make a neuron produce a real valued output by using an activation function. 


```python
Image(filename='./images/neuron.png')
```

In the neuron depicted above $a_1$,$a_2$ and $a_3$ are inputs and $w1$,$w2$ and $w3$ are weights associated with the input. The weighted combination of input $w^Ta$ is fed to an activation function. Initially the weights are randomly initialised and training a neural network is the process by which the weights are determined. Once the network is trained, whenever a input data is fed, computations are performed according to the flow of network (feedforward computation) and output is generated. 

## Linear regression

Linear regression is the 'hello world' of machine learning. In linear regression output is a linear combination of inputs.For the above network ,<br><br>
                         $$  h(x,w) = w_1*x_1 + w_2 * x_2 + w_3 * x_3 $$ <br>
                         
During training the inputs and outputs are fixed and we are required to learn the weights  that best fits our training data.  At any point of time, we need a measure to evaluate whether our solution is closer to the actual solution. Since all we have is some data and not the actual function, we use our data to estimate how well the neural network is performing. This measure is called as loss function which compares the actual output and predicted output and produces a real valued output. While there are a plenty of choices for loss functions , for this example lets assume the sum of squares loss function.

$$L(w) = \sum_i (h(x_i,w) - y) ^2$$ <br> 

During an iteration in training, we feed our data through the network,compute the loss and alter our weights in a way that the loss in successive iteration is less than the current loss.  Gradient descent is the algorithm which does this. We find how a small change in each weight affects the loss function (Derivative!) and change the weights in that direction which lowers the loss function. We repeat this for all our training data , one iteration over our training data is called as epoch.
