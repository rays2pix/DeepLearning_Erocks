## Function Approximation with Neural networks


One way to look at Neural Networks with fully-connected layers is that they define a family of functions that are parameterized by the weights of the network.Is it possible to model any function using neural network? Well theoritically yes! But lets give it a try.

```python
from nn_net import *
import matplotlib.pyplot as plot
import numpy as np
%matplotlib inline
```

Lets train our neural network to learn sine function.

```python
 
n=2000
X=np.linspace(0,3*np.pi,num=n)
X.shape=(n,1)
y=np.sin(X)

plot.plot(X,y,linewidth=2)
plot.show()
```

Below we create a neural network with one hidden layer. Lets experiment with this and analyze its predictive power.

```python
net_size = [1,30,1]
rate=0.05
param=((net_size[0],0,0),(net_size[1], expit, logistic_prime),(net_size[2],identity, identity_prime))
```

```python
n_iter =400
N=NeuralNetwork(X,y,param)
N.train(n_iter, learning_rate=rate)
```

```python
predictions=[]
predictions.append([rate,N.predict(X)])
    
fig, ax=plot.subplots(1,1)
    
ax.plot(X,y, label='Sine', linewidth=2, color='black')
for data in predictions:
    ax.plot(X,data[1],label="Prediction")
    ax.legend()
plot.show()
```

Lets try by increasing the number of hidden layers.

```python

n_iter=200
net_size = [1,3,3,1]
param=((net_size[0],0,0),(net_size[1], expit, logistic_prime),(net_size[2], expit, logistic_prime),(net_size[3],identity, identity_prime))
rate=0.01 #0.001,0.1,0.14

#net_size = [1,20,20,1]
#param=((net_size[0],0,0),(net_size[1], expit, logistic_prime),(net_size[2], expit, logistic_prime),(net_size[3],identity, identity_prime))
#rate=0.1 #0.001,0.1,0.14


N=NeuralNetwork(X,y,param)
N.train(40, learning_rate=rate)
```

```python
predictions=[]
predictions.append([rate,N.predict(X)])
    
fig, ax=plot.subplots(1,1)
    
ax.plot(X,y, label='Sine', linewidth=2, color='black')
for data in predictions:
    ax.plot(X,data[1],label="Predictions")
    ax.legend()
plot.show()
```

So provided enough number of hidden layers and appropriate parameters,we can approximate any continuous functions.
